# 矩陣基本運算

使用numpy套件的array，輸入矩陣a

$a=\begin{bmatrix}
1 & 2 & 3 \\
3 & 4 & 5
\end{bmatrix}
$

$
b=\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{pmatrix}
$

```
import numpy as np

a = np.array([[1, 2, 3],
              [3, 4, 5]])
print(a)

b=np.array([2,3,5],
           [4,5,6])
```

輸入矩陣b，並且使用b矩陣shape屬性，查詢其矩陣大小

$b=\begin{bmatrix}
2 & 1 & 3 \\
4 & 2 & 5
\end{bmatrix}
$



```
b=np.array([[2, 1, 3],
            [4, 2, 5]])
print(b)
print(b.shape)
```

## 轉置矩陣

使用b.T語法，求得b矩陣的轉置矩陣，存入c矩陣

$c=b'=\begin{bmatrix}
2 & 4 \\
1 & 2 \\
3 & 5
\end{bmatrix}
$

```
c = b.T
print(c)
```

## 矩陣加/減法
$d=a+b=
\begin{bmatrix}
1 & 2 & 3 \\
3 & 4 & 5
\end{bmatrix}+
\begin{bmatrix}
2 & 1 & 3 \\
4 & 2 & 5
\end{bmatrix}
=\begin{bmatrix}
3 & 3 & 6 \\
7 & 6 & 10
\end{bmatrix}
$

$
\begin{bmatrix}
1 & 2 & 3 \\
3 & 4 & 5
\end{bmatrix}-
\begin{bmatrix}
2 & 1 & 3 \\
4 & 2 & 5
\end{bmatrix}
=\begin{bmatrix}
-1 & 1 & 0 \\
-1 & 2 & 0
\end{bmatrix}
$

```
print('a+b')
d=a+b
print(d)

print('a-b')
print(a-b)
```

## 矩陣乘法(scalar product)
每一個矩陣元素相乘

$e=a*b=
\begin{bmatrix}
1 & 2 & 3 \\
3 & 4 & 5
\end{bmatrix} *
\begin{bmatrix}
2 & 1 & 3 \\
4 & 2 & 5
\end{bmatrix}
=\begin{bmatrix}
2 & 2 & 9 \\
12 & 8 & 25
\end{bmatrix}
$


```
e=a*b
print('a*b=')
print(e)
```

## 矩陣內積(dot product)
矩陣乘法

$e=a \times c=
\begin{bmatrix}
1 & 2 & 3 \\
3 & 4 & 5
\end{bmatrix} \times
\begin{bmatrix}
2 & 4 \\
1 & 2 \\
3 & 5 
\end{bmatrix}
=\begin{bmatrix}
13 & 23 \\
25 & 45
\end{bmatrix}
$


```
g=np.dot(a,c)
print(g)
```

```
from numpy.linalg import inv

Q=np.dot(a,a.T)
print('A*At')
print(Q)

N=inv(Q)
print('inv(A*At)')
print(N)
```

$
A = \begin{bmatrix} 
    a_{11} & a_{12} & \dots \\
    \vdots & \ddots & \\
    a_{k1} &        & a_{kk} 
    \end{bmatrix}
